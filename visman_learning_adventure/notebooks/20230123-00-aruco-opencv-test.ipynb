We're going to build out a texture for an aruco-tag, so that we can test out the pattern on a box in simulation.

- https://mecaruco2.readthedocs.io/en/latest/notebooks_rst/Aruco/aruco_basics.html

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt


def draw_aruco_grid(rows, cols, figsize=(10, 10)):
    # https://docs.opencv.org/3.4/d9/d6a/group__aruco.html#gac84398a9ed9dd01306592dd616c2c975
    aruco_dict = cv.aruco.Dictionary_get(cv.aruco.DICT_4X4_50)

    # 3x3 subplot
    fig, axes = plt.subplots(rows, cols, figsize=figsize)
    for i, ax in enumerate(axes.flatten()):
        img = cv.aruco.drawMarker(aruco_dict, i, 200)
        ax.imshow(img, cmap="gray", interpolation="nearest")
        ax.axis("off")


draw_aruco_grid(4, 4, figsize=(3, 3))
plt.show()
```

```python
draw_aruco_grid(4, 4, figsize=(10, 10))
plt.savefig(
    "../models/tabletop_aruco_test/materials/textures/grid.png",
    bbox_inches="tight",
    pad_inches=0,
)
```
